```python
import keras 
import tensorflow as tf

from keras.datasets import mnist

(x_train, y_train), (x_test, y_test) = mnist.load_data()

x_train.shape
x_train = x_train.reshape(-1, 28, 28, 1).astype('float32')/255.0
x_test  = x_test.reshape(-1, 28, 28, 1).astype('float32')/255.0 # this makes sense?
# x_train.shape

from keras.models import Sequential
from keras.layers import Dense, Conv2D, Flatten
model = keras.Sequential()

model.add(Conv2D(64, 3, input_shape = (28, 28, 1), padding = 'same', activation = 'relu'))

model.add(Conv2D(128, 3, activation = 'relu', padding = 'same'))

model.add(Flatten())

model.add(Dense(10))

model.compile(optimizer =  tf.keras.optimizers.Adam(), loss = tf.keras.losses.SparseCategoricalCrossentropy(from_logits = True), metrics = ['accuracy'])
# need a good enough hardware support if you want to train this data.
# x_train = x_train[0:3000]
# y_train = y_train[0:3000]
model.fit(x_train, y_train, epochs = 10)
```
